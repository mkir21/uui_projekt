```python
def ucs(graph, start, goal, heuristic=None):
    visited = set()
    queue = [(start, 0, [start])]

    while queue:
        queue.sort(key=lambda x: x[1])
        node, cost, path = queue.pop(0)

        if node in visited:
            continue

        visited.add(node)

        if node == goal:
            return (cost, path)

        for child, child_cost in graph.get(node, []):
            if child not in visited:
                queue.append((child, cost + child_cost, path + [child]))

    return None
```

```python
def astar(graph, start, goal, heuristic):
    node_info = {node: {'g': float('inf'), 'f': float('inf'), 'parent': None, 'visited': False} for node in graph}

    node_info[start]['g'] = 0
    node_info[start]['f'] = heuristic[start]
    node_info[start]['visited'] = True

    S = {start}

    while S:
        u = min(S, key=lambda node: node_info[node]['f'])

        if u == goal:
            path = []
            while u is not None:
                path.insert(0, u)
                u = node_info[u]['parent']
            return (node_info[goal]['g'], path)

        S.remove(u)

        for w, t in graph.get(u, []):
            gt = node_info[u]['g'] + t
            if gt < node_info[w]['g']:
                node_info[w]['parent'] = u
                node_info[w]['g'] = gt
                node_info[w]['f'] = gt + heuristic[w]

                if not node_info[w]['visited']:
                    S.add(w)
                    node_info[w]['visited'] = True

    print("A* could not find a path.")
    return None

heuristic = {
    label: sum(cvor[1] for cvor in children) for label, children in graf.items()
}
```
